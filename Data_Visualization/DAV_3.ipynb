```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/final_vg.csv')
df.head(2)
```

```python
# Extracting the top 3 publisher Genre and platform
top_3_platform = df['Platform'].value_counts()[:3].index
top_3_Genre = df['Genre'].value_counts()[:3].index
top_3_Publisher = df['Publisher'].value_counts()[:3].index
# top_3_platform = df[]
```

```python
top_3_filtered_data = df[(df['Platform'].isin(top_3_platform))&(df['Genre'].isin(top_3_Genre))&(df['Publisher'].isin(top_3_Publisher))]
```

```python
top_3_filtered_data.shape
```

```python
top_3_filtered_data.head()
```

```python
# NN Data
sns.scatterplot(x = 'NA_Sales', y = 'EU_Sales', data=top_3_filtered_data)
plt.title('NA vs EU Sales')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('NA_Sales', fontsize = 15)
plt.ylabel('EU_Sales', fontsize = 15)

```

```python
#NNC
sns.scatterplot(x = 'NA_Sales', y = 'EU_Sales',hue= 'Publisher', data=top_3_filtered_data)
plt.title('NA vs EU Sales')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('NA_Sales', fontsize = 15)
plt.ylabel('EU_Sales', fontsize = 15)
```

```python
#CN

sns.boxplot(x = 'Publisher', y ='Global_Sales', data = top_3_filtered_data)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Publisher wise global_Sales')
plt.show()
```

```python
#CCN
sns.boxplot(x = 'Publisher', y ='Global_Sales',hue ='Genre', data = top_3_filtered_data)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.xlabel('Publisher')
plt.ylabel('Global Sales')
plt.title('Publisher wise global_Sales')
plt.show()
```

```python
#Quiz1
fig = plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data = top_3_filtered_data)

plt.subplot(2,3,6)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top_3_filtered_data, color = 'r')

fig.suptitle('Main title')
```

```python
fig = plt.figure(figsize=(15,10))
plt.subplot(2,1,1)
sns.scatterplot(x='NA_Sales', y='EU_Sales', data = top_3_filtered_data)

plt.subplot(2,3,6)
sns.scatterplot(x='NA_Sales', y='JP_Sales', data=top_3_filtered_data, color = 'r')

fig.suptitle('Main title')
```

```python
#NNN
#I want a plot: which represents the
# relation between NA and JP sales, based on Ranking of games

sns.scatterplot(x = 'NA_Sales', y = 'JP_Sales',size = 'Rank', data = top_3_filtered_data)
plt.xlabel("NA Sales")
plt.ylabel("JP Sales")
plt.title("NA vs JP Sales, based on Ranking of the games", fontsize = 12)
plt.show()
```

```python
sns.jointplot(x = "NA_Sales", y = "EU_Sales", hue= "Genre", data = top_3_filtered_data)
plt.show()
```

```python
sns.pairplot(data = top_3_filtered_data, hue = "Genre")
plt.show()
```
