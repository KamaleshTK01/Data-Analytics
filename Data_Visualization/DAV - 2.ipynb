```python
top_3_platform = df['Platform'].value_counts().index[:3]
top_3_genre = df['Genre'].value_counts().index[:3]
top_3_pub = df['Publisher'].value_counts().index[:3]
```

```python
Top_3_df = df[(df['Platform'].isin(top_3_platform)) & (df['Genre'].isin(top_3_genre)) & (df['Publisher'].isin(top_3_pub))]
Top_3_df.head()
```
